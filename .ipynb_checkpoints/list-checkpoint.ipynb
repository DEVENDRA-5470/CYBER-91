```python
stu_marks=[45,89,98,45,65,44,33]
# 1. find highest marks
# 2. find lowest marks
# 3. filter marks gt 55

```

```python
# 1. find highest marks
stu_marks=[45,89,98,45,65,44,33]
high=stu_marks[0]
for i in stu_marks:
    if i > high:
        high=i
print(high)
```

```python
# 3. filter marks gt 55
stu_marks=[45,89,98,45,65,44,33]
gt=[]

for i in stu_marks:
    if i >= 55:
        gt.append(i)
print(gt)
```

```python
# list comprehension
# it use to write for loop inside list.
# when you want fast and less complex logic.
# it is fast than writing loop in traditional way.
```

```python
l=[]
for i in range(1,10):
    if i%2==0:
        l.append(i)
    elif i%3==0:
        l.append("yes 3")
    else:
        l.append("odd")
print(l)
```

```python
l=[i for i in range(1,10) if i%2==0]
print(l)
```

```python
l=[i if i%2==0 else "odd" for i in range(1,10) ]
print(l)
```

```python
zeros=[]
for i in range(1,10):
    zeros.append(0)
print(zeros)
```

```python
zeros=[i*0 for i in range(1,10)]
print(zeros)
```

```python
# nested list
nested=[1,22,33,44,['aman',"dev","shivam","kamal",[True,False],[23,44,55]]]
print(nested[4][4][0])
```

```python
nest=[[i for i in range(6)] for _ in range(6)]
# print(nest)
for i in nest:
    print(i)
```

```python
# list transform.
stock=[10,22,33,4,50,20]
new_stock=[i+2 for i in stock]
print(new_stock)
```

```python
# partition use
marks=[23,78,33,45,66,55,54,56,79,88,23,45]
paas=[i for i in marks if i>=50 ]
fail=[i for i in marks if i<=50 ]
print(paas)
print(fail)
```

```python
marks=[23,78,33,45,66,55,54,56,79,88,23,45]
tagged=["pass:"+str(i) if i>=50 else "fail:"+str(i) for i in marks]
print(tagged)
```

```python
marks=[23,78,33,45,66,55,54,56,79,88,23,45]
partition=[['pass']+[i for i in marks if i>=50],['fail']+[i for i in marks if i<=50]]
print(partition)
```
