```python
# numbers
# Type conversion(automatic) and type casting(forcefully).
# input() : it returns string type data by default.

num1=int(input("Enter your number 1st : "))
num2=int(input("Enter your number 2nd : "))
print(type(num1),type(num2))
res=num1+num2
print(f"Sum = {res}")
```

```python
# float
num1=float(input("Enter your number 1st : "))
num2=float(input("Enter your number 2nd : "))
print(type(num1),type(num2))
res=num1+num2
print(f"Sum = {res}")
```
