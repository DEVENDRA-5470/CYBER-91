```python
# 3.Logical operator (and,or,not)
and : return True if all condition are becomes true ,otherwise false
or : return True if any condition are becomes true ,otherwise false
not : its reverse the result.
```

```python
# and 
num1=30
num2=40
res=(num1<num2) and (num1 > 20) and (34>33) and (num1!=num2) and (num1==num2)
print(res)
```

```python
# or
num1=30
num2=40
res=(num1<num2) or (num1 > 20) or (34>33) or (num1!=num2) or (num1==num2)
print(res)
```

```python
# not
num1=30
num2=40
res=not((num1<num2) or (num1 > 20) or (34>33) or (num1!=num2) or (num1==num2))
print(res)
```
