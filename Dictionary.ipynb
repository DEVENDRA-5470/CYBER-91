```python
# 1.Defination and properites
# Dictionary is a data sturcture used to store value in key:value form.
# Each key:value combination becomes an item of the dictionary.
# Dictionary are mutable in nature.
# keys of a dictionary must be unique.
# key can not be a list,set.
# Dictionary defined by using {} and using dict() method.
# Value of a dictionary can be any data type.
# Dictionary are not support indexing in place we use keys.
```

```python
# 3.update/modification 
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
stu_marks['aman']=90
stu_marks["rohit"]=200
print(stu_marks)
```

```python
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
print(stu_marks['rohan'][0])
```

```python
# 5.Traversing 
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
for i,j in stu_marks.items():
    print(i,j)

print("-------")
for i in stu_marks.keys():
    print(i)

print("-------")
for i in stu_marks.values():
    print(i)
```

```python
# 6.built-in methods.
# get(keys,default)
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
res1=stu_marks.get('dev',"N/A")
print(res1)

# res1=stu_marks['dev']
# print(res1)
```

```python
# items()
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
res1=stu_marks.items()
print(res1)
```

```python
# values()
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
res1=stu_marks.values()
print(res1)
```

```python
# keys()
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
res1=stu_marks.keys()
print(res1)
```

```python
# len()
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
res1=len(stu_marks)
print(res1)
```

```python
# pop()
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
res1=stu_marks.pop('aman')
print(res1)
print(stu_marks)
```

```python
# popitem()
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
res1=stu_marks.popitem()
print(res1)
print(stu_marks)
```

```python
# update()
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
stu_marks.update({"pankaj":28,"danish":70,"aman":800})
print(stu_marks)
```

```python
# clear()
stu_marks={'aman':80,"kamal":90,"nitesh":60,"rohan":[100,80]}
stu_marks.clear()
print(stu_marks)
```
