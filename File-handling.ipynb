```python
# Creating file.
try:
    file=open("second.txt","w")
    file.write("Hello i am your first content in the file")
    file.flush()
    file.close()
    print("Done file created...")
except Exception as e:
    print(e)
finally:
    print("Thanks for visiting...")
```

```python
data=["Dev","kamal"]
with open("hello.txt","w+") as file:
    file.writelines(data)
    file.seek(0)
    read=file.read()
    print(read)
```

```python
with open("hello.txt","r") as file:
    read=file.read()
    print(read)
```

```python
with open("hello.txt","r+") as file:
    file.seek(0)
    read=file.read()
    file.write("this is new data")
    print(read)
```

```python
Name = "aman"
Roll = "45"
mobile = "00000000"
Address = "Delhi"

with open('student.txt',"w") as file:
    file.write(f'Name : {Name}')
    file.write("\n")
    file.write(f'Roll : {Roll}')
    file.write("\n")
    file.write(f'mobile : {mobile}')
    file.write("\n")
    file.write(f'Address : {Address}')
print("File written...")
```

```python
import random
import string
with open("coupon.txt","w") as file:
    for i in range(10):
        coupon="DEV-"+"".join([random.choice(string.digits+string.ascii_uppercase)
           for i in range(10)])
        file.write(coupon+"\n")
print("File creation done..")
```
