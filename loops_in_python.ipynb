```python
num1=2
num2=20
s=0
odd_sum=0
even_sum=0
mul=1
for e in range(num1,num2+1):
    mul*=e
    s=s+e # standard
    if e%2 !=0 :
        odd_sum+=e # shorthand
    else:
        even_sum+e
        
print(f"Sum : {s}")
print(f"Sum of odds : {odd_sum}")
print(f"Sum of even: {even_sum}")
print(f"Product : {mul}")
# sum of only odd numbers.
```

```python
for i in range(8):
    for j in range(8):
        if (i+j)%2==0:
            print("⬜",end="")
        else:
            print("⬛",end="")
    print()
```

```python
n=8
for i in range(n):
    for j in range(n):
        if i==0 or i==n-1 or j==0 or j==n-1:
            print("⬛",end="")
        else:
            print("⬜",end="")
    print()
```
